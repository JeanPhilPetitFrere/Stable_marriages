## Stable Mariages


```python
#man that are alone
lonely_men = []

#keeps track of the people that may end up together
fleeting_engagement = {}

def check_preference(woman,man,men,women):
    n1 = 0
    n2 = 0
    #returns the index of the current suitor
    for x in range(len(women[woman])):
        if women[woman][x] ==  fleeting_engagement[woman]:
            n1 = x
    #returns the index of the possible suitor
    for x in range(len(women[woman])):
        if women[woman][x] == man:
            n2 = x
    #if the possible suitor is preferred it replaces the current suitor by him        
    if n1 > n2:
        fleeting_engagement[woman] = women[woman][n2]
        lonely_men.remove(women[woman][n2])
        lonely_men.append(women[woman][n1])
        
        n1 *= 0
        n2 *= 0
    #if the current suitor is better it keeps the same suitor
    if n2 > n1:
        fleeting_engagement[woman] = women[woman][n1]


def ask_her_out(man,men,women):
    #goes through the prefer list of a man and ask's the lady out 
    for x in range(len(men[man])):
        if men[man][x] not in fleeting_engagement:
            num = men[man][x]
            fleeting_engagement[num] = man
            lonely_men.remove(man)
            break #breaks if he get's engaged
        else:
            check_preference(men[man][x],man,men,women)

def matchmaking(men=dict,women=dict):#takes in dict's with key -->id/name of applicants, value -->list of preferences
    #populates the lonely_men list
    fleeting_engagement.clear()#clear the fleeting engagement dict if you are using the algorithm more than once
    for man in men:
        lonely_men.append(man)
    #will stop our algorithm once their is no more lonely_men
    while len(lonely_men) > 0:
        for man in lonely_men:
            ask_her_out(man,men, women)
    #returns a dict with the couples
    return fleeting_engagement 

        
    
```

```python
#woman and their preferred match
match_woman = {
    0 : [0,1,2,3],
    1 : [2,3,0,1],
    2 : [1,0,3,2],
    3 : [3,2,1,0]
}
#man and their preferred match
match_man = {
    0 : [0,1,2,3],
    1 : [2,3,0,1],
    2 : [1,0,3,2],
    3 : [3,2,1,0]
}
matchmaking(match_man,match_woman)
```
